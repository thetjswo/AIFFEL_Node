## 계단 함수와 시그모이드 함수의 차이점
* 계단 함수: 출력 값이 0과 1로만 이루어진 반듯한 계단 모양을 그리는 함수
* 시그모이드 함수: 출력 값이 0~1 사이의 값으로 이루어진 부드러운 곡선을 그리는 함수

## 계단 함수와 시그모이드 함수의 공통점
* 입력이 작을 땐 출력이 0에 가까워지고, 입력이 클 땐 출력이 1에 가까워지는 구조
* 비선형 함수
    * 선형 함수: 입력된 값의 상수 배만큼의 값이 출력되는 함수(한 개의 직선으로 표현된다.)
    * 비선형 함수: 직선 1개로는 그릴 수 없는 함수
    
### 선형 함수를 활성화 함수로 쓰지 않는 이유
* 신경망의 층을 깊게하는 의미가 없다.
> $h(x) = cx$를 활성화 함수로 사용하는 3개층의 신경망을 만들어 보자.  
$y(x) = h(h(h(x)))$ 표현할 수 있고, 이 계산은 $y(x) = c*c*c*x$와 같다.  
$a = c^3$이라고 정의하고 $y(x) = ax$라고 하면 은닉층이 없는 네트워크로 표현할 수 있다.  

## ReLU 함수

* ReLU함수는 입력값이 0을 넘으면 그대로 출력하고, 0과 같거나 0보다 작으면 0을 출력하는 함수이다.
* 수식으로는 다음과 같이 쓸 수 있다.
$h(x) =
\begin{cases}
x (x \gt 0)\\
0 (x \leq 0)
\end{cases}
$

```python
def relu(x):
    return np.maximum(0,x) #입력되는 두 값 중 큰 값을 선택하여 반환

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-1, 6)
plt.legend('step')

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-1, 6)
plt.legend('sigmoid')

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1, 6)
plt.legend('relu')

plt.show()
```

## softmax function

$$y_k = {exp(a_k) \over \displaystyle\sum_{i=1}^{n}{exp({a_i})}}$$
* $exp(x)$는 $e^x$을 뜻하는 지수 함수(exponential function)
* $n$은 출력층의 뉴런 수
* $y_k$는 출력층의 뉴런 중 $k$번째 출력

소프트맥스 함수의 분자는 입력 신호 $a_k$의 지수 함수, 분모는 모든 입력 신호의 지수 함수의 합으로 구성된다.

```python
a = np.array([0.3, 2.9, 4.0])
exp_a = np.exp(a)
print(exp_a)

sum_exp_a = np.sum(exp_a)
print(sum_exp_a)

y = exp_a / sum_exp_a
print(y)
```

```python
def init_network():
    network = {}
    network['W1'] = np.array([[0.1, 0.3, 0.5],[0.2, 0.4, 0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network['b3'] = np.array([0.1, 0.2])
    
    return network

def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    
    a3 = np.dot(z2, W3) + b3
    y = identity_function(a3)
    
    return y

def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    
    y = exp_a / sum_exp_a
    
    return y

network = init_network()
x = np.array([1.0, 0.5])
a = forward(network, x)
y = softmax(a)
print(y)
```

### 소프트맥스 함수 사용 시, 오버플로우에 대비하는 방법

$$y_k = {exp(a_k) \over \displaystyle\sum_{i=1}^{n}{exp({a_i})}} = {C exp(a_k \over C \displaystyle\sum_{i=1}^{n}{exp(a_i)}} = {{exp(a_k + \log C)} \over \displaystyle\sum_{i=1}^{n}{exp(a_i + \log C)}} = {{exp(a_k + C`)} \over \displaystyle\sum_{i=1}^{n}{exp(a_i + C`)}}$$

```python
a = np.array([[1],[2],[3]])
print(a.shape)
b = np.array([[1,2,3]])
np.dot(a,b)
```

```python
a = np.array([1010, 1000, 990])
res = np.exp(a) / np.sum(np.exp(a))
print(res)

c = np.max(a)
print(c)

np.exp(a - c) / np.sum(np.exp(a-c))
```

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    
    y = exp_a / sum_exp_a
    
    return y
```

```python
a = np.array([0.3, 2.9, 4.0])

y = softmax(a)
print(y)

np.sum(y)
```

* 소프트맥스 함수의 출력은 0과 1 사이의 실수이다.
* 소프트맥스 함수의 출력들의 합은 1이다.
> 소프트맥스 함수의 가장 중요한 성질이며, 이 성질 덕분에 출력을 **확률**로 해석할 수 있다.  
위의 코드를 확률로서 해석을 한다면, y[0]의 확률은 1.8%, y[1]의 확률은 24.5%, y[2]의 확률은 73.65%로 2번째 클래스가 가장 답에 근접하다 라고 해석할 수 있다.

* 신경망을 이용한 분류에서는 출력층에서의 소프트맥스 함수를 생략하기도 한다.
> 소프트맥스 함수는 단조증가 함수로 입력 값의 대소 관계가 소프트맥스 출력 결과에서도 유지된다.  
신경망을 이용한 분류는 각 클래스 중 가장 큰 출력을 내는 뉴련에 해당하는 클래스만 인식하기 때문에 소프트맥스 함수 사용이 의미가 없다.

### 신경망의 추론 처리
* MNIST Dataset을 가지고 구현할 신경망의 입력층 뉴런은 784개, 출력층의 뉴런은 10개로 구성한다.
* 입력층의 뉴런이 784개인 이유는 이미지의 크기가 28*28인 784이기 때문이다.
* 출력층의 뉴런이 10개인 이유는 이 문제가 0에서부터 9까지의 수를 구분하는 문제이기 때문이다.

```python
import pickle

def get_data():
    (X_train, y_train), (X_test, y_test) = load_mnist(flatten=True, normalize=True, one_hot_label=False)
    return X_test, y_test

def init_network():
    with open("sample_weight.pkl", 'rb') as f:
        network = pickle.load(f)

    return network

def predict(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)

    return y
```

```python
X, t = get_data()
network = init_network()

accuracy_cnt = 0
for i in range(len(X)):
    y = predict(network, X[i])
    p = np.argmax(y)
    if p == t[i]:
        accuracy_cnt += 1

print('Accuracy:'+str(float(accuracy_cnt) /  len(X)))
```

### 배치 처리
* 배치란? x[0]과 y[0]에는 0번째 추론 결과가 저장되는 것과 같이 하나로 묶은 입력 데이터를 배치라고 부른다.

```python
X, t = get_data()
network = init_network()

batch_size = 100
accuracy_cnt = 0
for i in range(0, len(X), batch_size):
    X_batch = X[i:i+batch_size]
    y_batch = predict(network, X_batch)
    p = np.argmax(y_batch, axis=1)
    accuracy_cnt  += np.sum(p == t[i:i+batch_size])

print('Accuracy:'+str(float(accuracy_cnt) /  len(X)))
```
