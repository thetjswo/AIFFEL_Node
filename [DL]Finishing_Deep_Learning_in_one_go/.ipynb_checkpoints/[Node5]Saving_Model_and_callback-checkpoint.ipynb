```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import models, layers, utils
```

```python
from tensorflow.keras.datasets import mnist
from sklearn.model_selection import train_test_split

(x_train_full, y_train_full), (x_test, y_test) = mnist.load_data(path='mnist.npz')

x_train, x_val, y_train, y_val = train_test_split(x_train_full, y_train_full,
                                                  test_size=0.3,
                                                  random_state=123)

print(f"전체 학습 데이터: {x_train_full.shape}  레이블: {y_train_full.shape}")
print(f"학습용 데이터: {x_train.shape}  레이블: {y_train.shape}")
print(f"검증용 데이터: {x_val.shape}  레이블: {y_val.shape}")
print(f"테스트용 데이터: {x_test.shape}  레이블: {y_test.shape}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

num_sample = 6
random_idxs = np.random.randint(60000, size=num_sample)

plt.figure(figsize=(15,15))
for i, idx in enumerate(random_idxs):
    img = x_train_full[idx, :]
    label = y_train_full[idx]

    plt.subplot(1, len(random_idxs), i+1)
    plt.axis('off')
    plt.title(f'Index: {idx}, Label: {label}')
    plt.imshow(img)
```

```python
x_train = x_train / 255.
x_val = x_val / 255.
x_test = x_test / 255.

y_train = utils.to_categorical(y_train)
y_val = utils.to_categorical(y_val)
y_test = utils.to_categorical(y_test)

print(y_train[0])
print(y_val[0])
print(y_test[0])
```

```python
model = models.Sequential()
model.add(keras.Input(shape=(28, 28), name='input'))
model.add(layers.Flatten(input_shape=[28, 28], name='flatten'))
model.add(layers.Dense(100, activation='relu', name='dense1'))
model.add(layers.Dense(64, activation='relu', name='dense2'))
model.add(layers.Dense(32, activation='relu', name='dense3'))
model.add(layers.Dense(10, activation='softmax', name='output'))
model.summary()
```

```python
utils.plot_model(model)
```

```python
utils.plot_model(model, show_shapes=True)
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train,
                    epochs=50,
                    batch_size=128,
                    validation_data=(x_val, y_val))
```

```python
history.history.keys()
```

```python
history_dict = history.history

loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss) + 1)
fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, color='blue', label='train_loss')
ax1.plot(epochs, val_loss, color='red', label='val_loss')
ax1.set_title('Train and Validation Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.grid()
ax1.legend()

accuracy = history_dict['accuracy']
val_accuracy = history_dict['val_accuracy']

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, accuracy, color='blue', label='train_accuracy')
ax2.plot(epochs, val_accuracy, color='red', label='val_accuracy')
ax2.set_title('Train and Validation Accuracy')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.grid()
ax2.legend()

plt.show()
```

```python
model.evaluate(x_test, y_test)
```

```python
pred_ys = model.predict(x_test)
print(pred_ys.shape)
print(pred_ys[0])
```

```python
arg_pred_y = np.argmax(pred_ys, axis=1)

plt.title(f'Predicted label: {arg_pred_y[0]}')
plt.imshow(x_test[0])
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import models, layers, utils
from tensorflow.keras.datasets import mnist
from sklearn.model_selection import train_test_split
# from tensorflow.keras.callbacks import ModelCheckpoint
from keras.callbacks import EarlyStopping, ModelCheckpoint


# 학습률 조정 함수 정의
def scheduler(epoch, learning_rate):
    if epoch < 10:
        return learning_rate
    else:
        return learning_rate * tf.math.exp(-0.1)

# 학습 데이터, 학습 레이블, 겁증 데이터, 겁증 레이블, 평가 데이터, 평가 레이블 생성
(x_train_full, y_train_full), (x_test, y_test) = mnist.load_data(path='mnist.npz')

x_train, x_val, y_train, y_val = train_test_split(x_train_full, y_train_full,
                                                  test_size=0.3,
                                                  random_state=123)

# 각 데이터 정규화 진행
x_train = x_train / 255.
x_val = x_val / 255.
x_test = x_test / 255.

y_train = utils.to_categorical(y_train)
y_val = utils.to_categorical(y_val)
y_test = utils.to_categorical(y_test)

# 모델 설계
model = models.Sequential()
model.add(layers.Input(shape=(28,28), name='Input_Layer'))
model.add(layers.Flatten(input_shape=[28,28], name='Flatten_Layer'))
model.add(layers.Dense(100, activation='relu', name='Dense_Layer1'))
model.add(layers.Dense(64, activation='relu', name='Dense_Layer2'))
model.add(layers.Dense(32, activation='relu', name='Dense_Layer3'))
model.add(layers.Dense(10, activation='softmax', name='Output_Layer'))
model.summary()

# 모델 컴파일
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# 설계된 모델 확인
utils.plot_model(model, show_shapes=True)

# 체크 포인트 기준 설정
check_point_cb = ModelCheckpoint('check_point_of_mnist_model.keras', save_best_only=True)
# 조기 종료 기준 설정
early_stopping_cb = EarlyStopping(monitor='val_loss', patience=10)
# 학습률 조정 기준 설정
lr_scheduler_cb = callbacks.LearningRateScheduler(scheduler)

# 모델 학습
history = model.fit(x_train, y_train, epochs=100, 
                    validation_data=(x_val, y_val), 
                    callbacks=[check_point_cb, early_stopping_cb, lr_scheduler_cb])


# 학습 - 검증 간의 정확도/손실률 그래프 그리기
history_dict = history.history

loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss)+1)
fig = plt.figure(figsize=(12,5))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, color='blue', label='train_loss')
ax1.plot(epochs, val_loss, color='red', label='val_loss')
ax1.set_title('Train and Validation Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.grid()
ax1.legend()

accuracy = history_dict['accuracy']
val_accuracy = history_dict['val_accuracy']

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, accuracy, color='blue', label='train_accuracy')
ax2.plot(epochs, val_accuracy, color='red', label='val_accuracy')
ax2.set_title('Train and Validation Accuracy')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.grid()
ax2.legend()

plt.show()

# 예측 결과 확인하기
predictions = model.predict(x_test)
print(predictions.shape)
print(predictions[0])

arg_pred_y = np.argmax(predictions, axis=1)

plt.title(f'Predicted label: {arg_pred_y[0]}')
plt.imshow(x_test[0])
plt.show()
```
